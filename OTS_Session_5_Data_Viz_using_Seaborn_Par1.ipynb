# Using seaborn library for creating

- Bar Plots
- Box plots
- Histograms
- density Plots
- Joint plots
- Scatterplot matrix


```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Barplot

```
titanic = sns.load_dataset("titanic")

print(titanic.shape)
print(titanic.info(), '\n')

titanic.head()
```

# Barplot for single variable

```
sns.catplot(x="sex", data=titanic, kind='count')
```

Giving Plot title

```
sns.catplot(x="sex", data=titanic, kind='count')
plt.title('barplot', fontname='Ubuntu', fontsize=14,
            fontstyle='italic', fontweight='bold')
```

##Stacked Barplot for categorical variables

Using seaborn.catplot(), we can create barplot for
one numeric and one categorical variables
however, we prefer using boxplot in this case 

```
sns.barplot(x='pclass', y='fare', data=titanic)

# sns.catplot(x='sex', y='age', data=titanic, kind='bar')
```

## Adding more variables

```
sns.catplot(x="sex", y="fare", hue="class", kind="bar", data=titanic)
```

# Boxplot
## Boxplot for one nuerical variable

```
# sns.catplot(y='age', data=titanic, kind='box')
sns.boxplot(x='age', data=titanic)
# shows the three quartile values of the distribution along with extreme values. The “whiskers” extend to points that lie within 1.5 IQRs of the 
# lower and upper quartile and the points outside these are considered as outliers.
```

##Box Plot for one numeric & one categorical variable

```
sns.catplot(x='sex', y="age", data=titanic, kind='box')
```

## Box plot for one numeric, two categorical variables

```
sns.catplot(x='sex', y='fare', hue='class', data=titanic, kind='box')

# pd.crosstab(titanic['sex'], titanic['class'])
# 122/(122+108+347)
# 94/(94+76+144)
```

##Violin Plots/Boxen Plots

```
sns.catplot(y='sex', x="age", data=titanic, kind='boxen')
```

```
sns.catplot(y='sex', x="age", data=titanic, kind='violin')
# combines a boxplot with the kernel density estimation, the quartile and whikser values from the boxplot are shown inside the violin
```

## More plots for categorical data

```
df = pd.read_csv('/content/drive/My Drive/OTS_Data/mtcars.csv')
print(df.info())
df.head()
```

```
sns.catplot(x='cyl', y='mpg', data=df)

# sns.catplot('cyl', 'mpg', data=df, kind='strip') #default setting is for stripplot
# It adjusts the positions of points on the categorical axis with a small amount of random “jitter”:
```

```
sns.catplot('cyl', 'mpg', data=df, jitter=False) # setting jitter=False
```

```
# Swarmplot/ Beeswarmplot
sns.catplot('cyl', 'mpg', data=df, kind='swarm')

# adjusts the points along the categorical axis using an algorithm that prevents them from overlapping
# it only works well for relatively small datasets
```

```
# We can add another dimension to a categorical plot by using a hue semantic. 
# (The categorical plots do not currently support size or style semantics)

sns.catplot('cyl', 'mpg', hue='gear', data=df)

# sns.catplot('gear', 'mpg', data=df)
```

# More plots for distribution

### Joint plots
### Hexbin Plots


```
sns.jointplot('wt', 'mpg', data=df)

# sns.distplot(df.wt)
```

```
#@title
#Hexbin Plots
# This plot shows the counts of observations that fall within hexagonal bins.
sns.jointplot('wt', 'mpg', data=df, kind='hex')
# https://cran.r-project.org/web/packages/hexbin/vignettes/hexagon_binning.pdf
```

```
# Bivariate distribution can also be visualized using the kernel density estimation.
sns.jointplot('wt', 'mpg', kind='kde', data=df)
```

```
# Adding layers to current plot
s = sns.jointplot('wt', 'mpg', kind='kde', data=df)
s.plot_joint(plt.scatter, c="r", s=30,  marker="*")
```

```
sns.pairplot(df)
```

```
sns.pairplot(df, hue='cyl')
```
